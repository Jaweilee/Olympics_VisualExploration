```python
def search_data_from_noc(b):
    result=[]   # List of known country matches
    unk_result=[] # List of unknown/missing codes
    for a in b:
        i= pycountry.countries.get(alpha_3=a[1])
        if i != None:
            result.append(i)
        else:
            unk_result.append(a)
    return result , unk_result

```

```python
def search_numeric_from_alpha3(b):
    for index,row in b.iterrows():
        try:
            n=pycountry.countries.get(alpha_3=row['alpha3code'])
            row['numeric']= n.numeric
        except AttributeError:
            pass
    return b['numeric']
 
```
