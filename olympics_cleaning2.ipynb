```python
import pandas as pd
import pycountry
```

```python
df= pd.read_csv('updated_olympics.csv',dtype={'numeric':'object'})
```

```python
df['medal'].unique()
#check the unique values for the medals
```

```python
df['medal'].fillna("NA", inplace = True)
#fill any NaN values
```

```python
df.columns
```

```python
gamesdrop= df[(df['season'] == 'Winter') ].index
```

```python
df.rename(columns={'alpha3code': 'countrycode'}, inplace=True)
```

```python
df.drop(gamesdrop, inplace=True)
df.shape
```

```python
df.columns
```

```python
df = df.drop(['Unnamed: 0','age','height','weight'], axis=1)
df.head()
#drop unneeded columns not going to be used for the visualization
```

```python
df = df.drop(df[df['year'] <1948].index)
#drop years before 1948 as I focus on Olympics after WWII
```

```python
df['team'].unique()
#find the unique Olympic team
```

```python
df['sport'].unique()
```

```python
df['games'].unique()
#check the games are only for Summer
```

```python
df.head(10)
```

```python
df['Bronze Medal'] = df['medal'].apply(lambda x: 1 if x in ['Bronze'] else 0)
df['Silver Medal'] = df['medal'].apply(lambda x: 1 if x in ['Silver'] else 0)
df['Gold Medal'] = df['medal'].apply(lambda x: 1 if x in ['Gold'] else 0)
#generate medal columns for each type
```

```python
df.head()
```

```python
#drop more unnecessary columns
df1 = df.drop(['name','sex','team','noc','games','season','city','event','medal','medal_count'], axis=1)
df1.head()
```

```python
df_agg=df1.groupby(['year','sport','countrycode','numeric']).agg({'id':'count','Bronze Medal':'sum','Silver Medal':'sum','Gold Medal':'sum' })
```

```python
df_agg=df_agg.reset_index()
```

```python
df_agg
```

```python
country_list=[]
for index,row in df_agg.iterrows():
        try:
            c=pycountry.countries.get(alpha_3=row['countrycode'])
            country= c.name
            country_list.append(country)
        except AttributeError:
            country_list.append('NaN')
```

```python
df_agg['country']=country_list
```

```python
new_order=['year','sport','country','countrycode','numeric','id','Bronze Medal','Silver Medal','Gold Medal']

df_agg=df_agg[new_order]
df_agg.rename(columns={'id': 'number_of_athletes'}, inplace=True)
df_agg.head()
```

```python
df_agg.shape
```

```python
df_agg.to_csv('updated_olympics_aggregation.csv',index=False)
```
